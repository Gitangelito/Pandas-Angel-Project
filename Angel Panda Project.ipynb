```python
#Age distribution chart count
sns.histplot(df['Age'])
plt.title("Age Distribution")
plt.show()
```

```python
#Gender distribution count
gender_counts = df['Gender'].value_counts() 
print(gender_counts)
```

```python
#location counts
location_counts = df['Location'].value_counts()
print(location_counts)
```

```python
grouped_data = df.groupby(['Gender', 'Location']).size().reset_index(name='Count')
pivot_table = grouped_data.pivot_table(index='Location', columns='Gender', values='Count', fill_value=0)

pivot_table.plot(kind='bar', figsize=(10, 6))
plt.title("Location Distribution by Gender")
plt.xlabel("Location")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
#Item popularity
item_popularity = df['Item Purchased'].value_counts()
print(item_popularity)
```

```python
item_popularity.plot(kind='bar')
plt.title("Item Popularity")
plt.xlabel("Item Purchased")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
item_popularity.plot(kind='barh')
plt.title("Item Popularity")
plt.xlabel("Count")
plt.ylabel("Item Purchased")
plt.show()
```

```python
# Define age ranges
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '61+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Group data by age group and item purchased
grouped_data = df.groupby(['Age Group', 'Item Purchased']).size().reset_index(name='Count')

# Create pivot table
pivot_table = grouped_data.pivot_table(index='Item Purchased', columns='Age Group', values='Count', fill_value=0)

# Create heatmap
sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title("Item Popularity by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Item Purchased")
plt.show()
```

```python
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '61+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Group data by age group and item purchased, calculate average spending
grouped_data = df.groupby(['Age Group', 'Item Purchased'])['Purchase Amount (USD)'].mean().reset_index(name='Average Spending')

# Create pivot table
pivot_table = grouped_data.pivot_table(index='Item Purchased', columns='Age Group', values='Average Spending', fill_value=0)

# Create heatmap
sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title("Average Spending by Age Group and Item")
plt.xlabel("Age Group")
plt.ylabel("Item Purchased")
plt.show()
```
